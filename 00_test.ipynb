### Create random tensors

Random tensors are tensors of some arbitrary numbers.

```python
from tensorflow._api.v2 import random
# Create two random tensors
random_1 = tf.random.Generator.from_seed(42)
random_1 = random_1.normal(shape=(3,2))
random_2 = tf.random.Generator.from_seed(42)
random_2 = random_2.normal(shape=(3,2))
random_1,random_2

# Are the equal?
random_1 == random_2
```

### Shuffle the order of elements in tensor

### Getting information from tensors

* Shape (number of elements)-- tensor.shape
* Rank (number of tensor dimensions)-- tensor.ndim
* Axis or dimension (particular dimension) -- tensor[0]
* Size (total number of items)-- tf.size(tensor)



```python
# Create a rank for tensor (4 dimension)

rank_4_tensor = tf.zeros([3,3,3,3])
rank_4_tensor
```

```python
rank_4_tensor.shape, rank_4_tensor.ndim, tf.size(rank_4_tensor)
```

```python
# Get various attribute of our tensor
print("Datatype of every element:", rank_4_tensor.dtype)
print("Number of dimensions:", rank_4_tensor.ndim)
print("Shape of tensor:", rank_4_tensor.shape)
print("Elements along the 0 axis:", rank_4_tensor.shape[0])
print("Elements along the last axis:", rank_4_tensor.shape[-1])
print("Total number of elements in a tensor:", tf.size(rank_4_tensor))
print("Total number of elements in a tensor in numpy:",tf.size(rank_4_tensor).numpy())
```

### Indexing tensors
Tensors can be indexed just like Python lists.

```python
# Get the first 2 elements of each dimension
rank_4_tensor[:2,:2,:2,:2]
```

```python
rank_4_tensor[:-1,:,:]
```

```python
# Get the first element from each dimension from each index except for the final one
rank_4_tensor[:1,:1,:1]
```

```python
from tensorflow.python.ops.gen_array_ops import rank
# Create a rank 2 tensor (2 dimensions)
rank_2_tensor = tf.constant([[10,7],[2,8]])#tf.zeros(shape=(2,2))
rank_2_tensor.ndim, rank_2_tensor.shape
```

```python
# Get the last item of each of our rank 2 tensor
rank_2_tensor[:,-1]
```

```python
# Add in extra dimension to our rank 2 tensor
rank_3_tensor = rank_2_tensor[...,tf.newaxis]
rank_2_tensor,rank_3_tensor
```

```python
#Alternative to tf.newaxise->expand_dims(input, axis)
tf.expand_dims(rank_2_tensor, axis=-1) # "-1" means expand the final axis
```

```python
tf.expand_dims(rank_2_tensor, axis=0) # expand the 0-axis
# same as
# rank_2_tensor[tf.newaxis,...]
```

**Matrix multiplication**

In machine learning, matrix multiplication is one of the most common tensor operations.

```python
# Matrix mulitiplication in TensorFlow
print(tensor)
tf.matmul(tensor,tensor)
```

```python
tensor*tensor
```

```python
tensor1 = tf.constant([[[1,2,5],[7,2,1],[3,3,3]]])
tensor2 = tf.constant([[[3,5],[6,7],[1,8]]])
tensor1, tensor2
```

```python
# Testing matmul using tensor1 and tensor2
tf.matmul(tensor1, tensor2)
```

```python
# Matrix multiplication with Python operator "@"
tensor @ tensor
```

```python
# Create two shape(3,2) tensors
X = tf.constant([[1,2],[3,4],[5,6]])
Y = tf.constant([[1,2],[3,4],[5,6]])
X,Y
```

```python
# Multiply them.
# wrong way: tf.matmul(X,Y), because their inner dimension does not match
tf.matmul(X,tf.reshape(Y, shape=(2,3)))
```

```python
# If we reshape X to (2,3), the shape of the result of mutiplication is (2,2)
tf.matmul(tf.reshape(X, shape=(2,3)),Y)
```

```python
# Use transpose
# perm represent the dimension of original matrix that new matrix to be
# e.g if the first dimension is unchanged, then the first value of perm is 0 (per=[0....])
X, tf.transpose(X), tf.reshape(X, shape=(2,3))
```

```python
# Transpose practice
Z = tf.constant([[[1,2],[3,4],[5,6]]])
Z, tf.transpose(Z, perm=[0,2,1])
```

**The dot product**

Matrix multiplication is also referred to as the dot product.

You can perform matrix multi using:
* `tf.matmul()`
* `tf.tensordot()`

```python
# Perform the dot product on X and Y (requires X or Y to be transposed)
tf.tensordot(tf.transpose(X),Y,axes=1)
```

```python
# Perform matrix multiplication between X and Y (transposed)
tf.matmul(X, tf.transpose(Y))
```

```python
# Perform matrix multiplication between X and Y (reshaped)
tf.matmul(X, tf.reshape(Y,(2,3)))
```

```python
# Y, reshaped Y and transposed Y
print(Y,"\n")

print(tf.reshape(Y, (2,3)),"\n")
print(tf.transpose(Y),"\n")
```

### Changing the datatype of a tensor

```python
# Create a new tensor with default datatype (float32)
B = tf.constant([1.7,7.4])
B.dtype
```

```python
C =  tf.constant([1,7])
C.dtype
```

```python
# Change from float32 to float16 (reduced precision)
D = tf.cast(B, dtype=tf.float16)
D, D.dtype
```

```python
# Change from int32 to float32
E = tf.cast(C, dtype=tf.float32)
E
```

```python
E_float16 = tf.cast(E, dtype=tf.float16)
E_float16
```

### Aggregating tensors
Aggregating: condensing them from multiple values down to a smaller amount of values.

```python
# Get the absolute values
D = tf.constant([-7,-10])
absolute_D = tf.abs(D)
D, absolute_D
```

Various forms of aggregation:
* minimum
* maximum
* mean
* sum

```python
# Create a random tensor with values between 0 and 100 with a size of 50
E = tf.constant(np.random.randint(0,100, size=50))
E
```

```python
tf.size(E), E.shape, E.ndim
```

```python
# Find the minimum
tf.reduce_min(E)
```

```python
# Find the maximum
tf.reduce_max(E)
```

```python
# Find the mean
tf.reduce_mean(E)
```

```python
# Find the sum
tf.reduce_sum(E)
```

**Exercise:** find the variance and standard deviation

```python
# variance
E_float32 = tf.cast(E, dtype=tf.float32)
tf.math.reduce_variance(E_float32)
```

```python
# standard deviation
tf.math.reduce_std(E_float32)
```

```python
# another method to get var. and std.
import tensorflow_probability as tfp
tfp.stats.variance(E)
```

```python
tfp.stats.stddev(E_float32)
```
